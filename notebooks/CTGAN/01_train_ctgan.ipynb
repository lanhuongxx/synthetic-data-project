```python
import pandas as pd

df = pd.read_csv(
    "../../data/processed/credit_default_clean.csv"
)

df.head()
```

```python
df.shape
```

```python
from sdv.metadata import SingleTableMetadata

metadata = SingleTableMetadata.load_from_json(
    "../../reports/metadata/credit_default_metadata.json"
)

metadata.to_dict()
```

```python
# khoi tao model CTGAN
from sdv.single_table import CTGANSynthesizer

synthesizer = CTGANSynthesizer(
    metadata,
    epochs=100,
    verbose=True
)
```

epochs=100 de test sau nang len 300, 500

```python
# Train model
synthesizer.fit(df)
```

```python
synthesizer.save(
    "../../models/ctgan_credit_default.pkl"
)
```

```python
# sinh du lieu gia
synthetic_data = synthesizer.sample(
    num_rows=30000
)

synthetic_data.head()
```

```python
synthetic_data.shape
```

```python
synthetic_data.to_csv(
    "../../data/synthetic/credit_default_synthetic.csv",
    index=False
)
```
