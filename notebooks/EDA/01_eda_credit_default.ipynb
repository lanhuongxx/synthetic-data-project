```python
# 8. Ve phan phoi tuoi cua khach hang
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.histplot(df["AGE"], bins=30, kde=True)
plt.title("Phan phoi tuoi cua khach hang")
plt.xlabel("Tuoi")
plt.ylabel("Tan suat")
plt.show()
```
