```python
# 1 load data
import pandas as pd

df = pd.read_excel(
    "../../data/raw/default-of-credit-card-clients.xls",
    header=1
)

df.head()
```

```python
# xoa cot ID
df.drop("ID", axis=1, inplace=True)
```

```python
df.columns
```

```python
# 3 Missing values
df.isnull().sum()

missing_pct = df.isnull().mean() * 100
missing_pct.sort_values(ascending=False)
```

```python
# 4 Duplicate records
df.duplicated().sum()
```

```python
# 5 doi ten cot target
df.rename(
    columns={
        "default payment next month":"TARGET_DEFAULT"
    },
    inplace=True
)
```

```python
df.columns
```

```python
# 6 Constant columns
constant_cols = [
    col
    for col in df.columns
    if df[col].nunique() == 1
]

constant_cols
```

```python
# 7 High Cardiinality
cardinality = df.nunique()

cardinality.sort_values(ascending=False)
```

```python
# 8 Metadata cho SDV
from sdv.metadata import SingleTableMetadata

metadata = SingleTableMetadata()

metadata.detect_from_dataframe(df)
```

```python
metadata.to_dict()
```

```python
for col in [
    "PAY_0",
    "PAY_2",
    "PAY_3",
    "PAY_4",
    "PAY_5",
    "PAY_6",
]:
    metadata.update_column(
        column_name=col,
        sdtype="categorical"
    )
```

```python
metadata.update_column(
    column_name="TARGET_DEFAULT",
    sdtype="categorical"
)
```

```python
metadata.to_dict()
```

```python
df.to_csv(
    "../../data/processed/credit_default_clean.csv",
    index=False
)
```
